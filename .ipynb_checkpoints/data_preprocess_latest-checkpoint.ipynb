```python
import numpy as np
import rasterio
import matplotlib.pyplot as plt
import pyproj
from skimage.restoration import denoise_nl_means, estimate_sigma
from skimage import exposure
from skimage import img_as_float, img_as_ubyte, io
from skimage.morphology import octagon
import skimage.morphology as morph
```

```python
#from rasterio import features
with rasterio.open('out_raster/25m_vari.tif') as dataset:
    vari = dataset.read(1).astype(np.float32)
    meta = dataset.meta
    nodata_value = meta['nodata']  # Extract the nodata value
```

```python
plt.hist(vari.flatten(), bins=100)
```

```python
footprint = octagon(5,5)
```

```python
vari_dilated = morph.dilation(vari, footprint)
```

```python
fig, (a1, a3) = plt.subplots(1,2, figsize=(12,4))
a1.hist(vari.flatten())
a3.hist(vari_dilated.flatten(), bins=100)
```

```python
fig, (a1, a2) = plt.subplots(1,2, figsize=(12,4))
a1.imshow(vari)
a2.imshow(vari_dilated)
```

```python
fig, (a1, a2) = plt.subplots(1,2, figsize=(12,4))
a1.hist(vari.flatten(), bins=100)
a2.hist(vari_dilated.flatten(), bins=100)
```

```python
# green farm
new_tiff_profile_green = dataset.profile  # Copy the profile from the original dataset
new_tiff_profile_green.update(
            dtype=rasterio.float32,  # Update the data type to match the VARI data
                        count=1,  # Only one band for VARI
                        compress='lzw',  # You can choose a compression method if needed
                        tiled=False,
                        blockysize=1,
                        nodata=0.0
)
new_crs_green = pyproj.CRS.from_epsg(32651)
new_tiff_profile_green.update(crs=new_crs_green)
#export data to file
with rasterio.open('out_raster/25m_vari_dilated.tif', 'w', **new_tiff_profile_green) as new_tiff_green:
                            new_tiff_green.write(vari_dilated, 1)
```
